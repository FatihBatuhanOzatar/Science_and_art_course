```python
import torch
from diffusers import AutoPipelineForText2Image
from PIL import Image

print("Stable Diffusion (SDXL) is loading...")
print("This may take a few minutes.")


pipe = AutoPipelineForText2Image.from_pretrained(
    "stabilityai/stable-diffusion-xl-base-1.0",
    torch_dtype=torch.float16,
    variant="fp16"
)

# send model to GPU
pipe = pipe.to("cuda")
print("Model succesfully loaded.")
prompt_1 = "A futuristic city in oil painting style."

print("Generating image...")

# generate image calling the 'pipe' function
image_1 = pipe(prompt=prompt_1).images[0]

print("Image generated. Saving...")
display(image_1)
image_1.save("futuristic_city.png")
```

```python
prompt_2 = "A derelict robot reclining against a wall, malfunctioning with visible sparks, semi-seated pose, industrial decay, chiaroscuro."
print("Generating image...")
image_2 = pipe(prompt=prompt_2).images[0]
image_2.save("robot_philosopher.png")
display(image_2)
```
