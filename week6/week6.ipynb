```python
%%writefile app.py
import streamlit as st
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline
import torch
import warnings

# Suppress warnings for a cleaner app interface
warnings.filterwarnings("ignore")

@st.cache_resource
def load_model():
    """
    Loads the Mistral 7B model and tokenizer one time.
    """
    model_id = "mistralai/Mistral-7B-Instruct-v0.2"
    tokenizer = AutoTokenizer.from_pretrained(model_id)
    model = AutoModelForCausalLM.from_pretrained(
        model_id,
        device_map="auto",
        torch_dtype=torch.float16,
        load_in_4bit=True
    )

    chatbot = pipeline(
        "text-generation",
        model=model,
        tokenizer=tokenizer,
        max_new_tokens=512,
        temperature=0.7
    )

    return tokenizer, chatbot

# Chatbot Response Function (from Week 5)
def get_response(question, tokenizer, chatbot):
    """
    Asks the model a question and returns only the answer.
    """
    messages = [{"role": "user", "content": question}]
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)
    output = chatbot(prompt)[0]['generated_text']
    answer = output.split("[/INST]")[-1].strip()
    return answer

# --- Streamlit Web Interface (The UI) ---
st.set_page_config(page_title="Mistral Chatbot", layout="wide")
st.title("ğŸ¤– Science & Art - Week 6 Chatbot")
st.markdown("This UI connects to the Mistral-7B model from Week 5.")
with st.spinner("Loading Mistral-7B model... This may take a few minutes."):
    tokenizer, chatbot = load_model()
st.success("Model loaded successfully!")
st.subheader("Ask the Chatbot a Question")
user_input = st.text_input("You:", "")
if user_input:
    with st.spinner("Just a sec..."):
        answer = get_response(user_input, tokenizer, chatbot)
        st.subheader("Mistral:")
        st.write(answer)
```
